# Tuples

A tuple is similar to list.

The diffrenece between the two is that we can't change the elements of tuple once it is assigned wheres in the list, elements can be changed.

A tuple is a sequence of immutable Python objects.

# Tuple creation

```python
# Empty tuple
t = ()
print(t)  

# Tuple having integers
t = (1, 2, 3)
print(t) 

# tuple with mixed datatypes
t = (1, "shivan", 3.4)
print(t)  

# nested tuple
t = ("shubham", "shivan", [8, 4, 6], (1, 2, 3))


print(t)
```

```python
#only parentesis is not enough

t = ('shivan')
type(t)
```

```python
#need a comma at the end.

t = ('shivan',)
type(t)
```

```python
# parentesis is optional

t = 'shivan',
print(type(t))

print(t)
```

```python
t = ('shivan','shubham')
type(t) 
```

# Accessing Elements in Tuple

```python
t = ('shivan', 'shubham', 'shashi', 'sanju')

print(t[1])
```

```python
# Negative index

print(t[-1])      # Print last element in a tuple
```

```python
# Nasted tuple

t = ('ABC', ('shubham', 'shashi', 'shivan'))

print(t[1])
```

```python
print(t[1][2])
```

```python
print(t[0][2])
```

# Slicing

```python
t = (1, 2, 3, 4, 5, 6, 7)

print(t[1:4])         # 0 to n-1

# print element from starting to 2nd last elements
print(t[:-2])

# print element from starting to end
print(t[:])
```

# Changing a Tuple

Unlike lists, tuples are immutable.

This means that elements of a tuple cannot be changed once it has been assigned. But, if the element is itself a mutable datatype like list, its nested items can be changed.

```python
# create a tuple
t = (0, 1, 2, 3, [6, 5])

t[2] = 'x'      # will get TypeError becuse tuple are immutable.
```

```python
t = (0, 1, 2, 3, [6, 5])
t[4][1] = 'shivan'
print(t)
```

```python
# COncatniting tuples

t = (1, 2, 3) + (4, 5, 6)
print(t)
```

```python
# Repet the elements in a tuple for a given number of itmes using the *.

t = (('shivan',) * 4)
print(t)
```

# Tuple Deletion

```python
# we cannot change the elements in a tuple.

# delete entire tuple using del keyword
t = (1, 2, 3, 4, 5, 6, 7, 8)

# delete entire tuple

del t
```

# Built in Function

# Tuple Length

```python
t =  (1, 2, 3, 4, 5, 6, 7, 8)

print(len(t))
```

# Tuple sort

```python
t = (4, 5, 3, 7, 6)

new_t = sorted(t) 
print(new_t) # Take elements in the tuple and return a new sorted list
             # (Does not sort the tuple itself).
```

```python
# Get the largest element in a tuple

t = (2, 5, 1, 6, 9)

print(max(t))
```

```python
# Get the smallest element in a tuple

print(min(t))
```

```python
# Get the sum element in a tuple

print(sum(t))
```
